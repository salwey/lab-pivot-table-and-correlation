#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1
# your answer here

#superstore.pivot_table(index= ['Category','Region'], values = ['Sales','Profit'], aggfunc = ['sum','mean'])
display(employee.pivot_table(index=['Department'], values = ['Salary'], aggfunc = ['mean']))
```

```python
# Method 2
# your answer here
#superstore.groupby(['Segment','Ship Mode Plus']).agg({'Segment':pd.Series.unique,'Ship Mode':pd.Series.unique,'Ship Mode Plus':pd.Series.unique})
#employee[['Department','Salary']].groupby(['Department']).agg({np.mean})
employee[['Department','Salary']].groupby(['Department']).agg({'Salary':np.mean})
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
# your answer here
display(employee.pivot_table(index=['Department','Title'], values = ['Salary'], aggfunc = ['mean']).reset_index())
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python
# your answer here
employee[['Title','Years','Salary']].groupby(['Title','Years']).agg({'Salary':np.mean}).reset_index()
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here
employee[['Department','Title','Salary','Name']].groupby(['Department','Title']).agg({'Salary':np.mean, 'Name':pd.Series.nunique}).reset_index()
# Confirmed name is unique...though perhaps better to count number of index instead?
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
display(employee[['Department','Title','Salary']].groupby(['Department','Title']).agg(['mean', 'min', 'max', 'std']).reset_index())
display(employee)

# We get some NaNs in our results, this appears to be when we only have a single row in that grouping. (Carlos, Ana and Pedro)
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# your answer here
employee[['Department','Title','Salary']].groupby(['Department','Title']).agg(['mean', 'min', 'max', 'std']).reset_index().fillna(0)
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here
#employee[['Department','Name','Salary']].groupby(['Department']).agg({'Name':pd.Series.nunique, 'Salary':sum}).reset_index()
employee[['Department','Name','Salary']].groupby(['Department']).agg({'Name':'size', 'Salary':'sum'}).reset_index()
# reading up, size can be used but in quotes and we can put it all in quotes like with the pivot_table examples seen..and this is an actual count rather than unique count...
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here
employee[['Department','Title','Years','Salary']].groupby(['Department','Title']).agg({'Years':sum, 'Salary':'mean'}).reset_index()
```
